```python
jesse = pd.read_csv("jesse_results.csv")
meta = pd.read_csv("./meta_strategy.ipynb/results_meta.csv")
qld = pd.read_csv("qld_simple_results.csv")
ot = pd.read_csv("results_old_tech.csv")
```

```python
fig = px.line(jesse["equity"])
fig.add_scatter(y=meta["new"], name ="Meta")
fig.add_scatter(y=qld["equity"], name = 'qld')
fig.add_scatter(y=ot["equity"], name ="Old Tech")

fig.update_layout(template='plotly_dark')
fig.show()
```
